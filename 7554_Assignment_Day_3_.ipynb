1. use IF ELSE and ELIF to write a program in python for your Report Cards

```python
"""
Grade A+ -> greater than 
Grade A -> greater than 75 and smaller than and equal to 90
Grade B+ -> greater than 65 and smaller than and equal to 75
Grade B -> greater than 55 and smaller than and equal to 65
Grade C+ -> greater than 45 and smaller than and equal to 55
Grade C -> greater than 36 and smaller than and equal to 45
Otherwise Fail
"""
# Ask the user to enter the mark.

student_mark = int(input('Enter the mark:  '))

if student_mark > 90:
    print("Grade A+")
elif student_mark > 75 and student_mark <= 90:
    print("Grade A")
elif student_mark > 65 and student_mark <= 75:
    print("Grade B+")
elif student_mark > 55 and student_mark <= 65:
    print("Grade B")
elif student_mark > 45 and student_mark <= 55:
    print("Grade C+")
elif student_mark > 36 and student_mark <= 45:
    print("Grade C")
else:
    print("Fail")
```

2. Use For Loop to Print Prime Numbers in between 1 to 1000 

```python
# Use For Loop to Print Prime Numbers in between 1 to 1000 
for num in range(1,1000):
    if num > 1:
        for i in range(2,num):
            if num % i == 0:
                print(f"{num} is not a prime number")
                break
        else:
            print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
            
```

3. Write a program for printing the tabels from 1,10 using Nested For Loop

```python
for i in range(1,11):
    for j in range(1,11):
        print(i, '*' , j, " = " , i*j)
    print(" ")
        
```
